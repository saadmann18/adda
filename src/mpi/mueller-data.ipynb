```python
import numpy as np
import pandas as pd
```

```python
import os
```

## Ellipsoid

```python
# Folder Path ellipsoid
dir_path = "/home/saad/adda/adda/src/mpi/ellipsoid"
resfile = []
resfolder = []
resdirpath = []
for (dir_path, dir_names, file_names) in os.walk(dir_path):
    resfile.extend(file_names)
    resfolder.extend(dir_names)
    resdirpath.append(dir_path)
```

```python
df = pd.DataFrame({'folder_path':[], 'Geometry':[], 
                   'degrees':[],
                   'S11': [], '-S21/S11': [], 
                   'y/x': [], 'z/x': [], 'orient': [], 
                   'label': []})
```

```python
df['folder_path'] = resdirpath[1:]
df['Geometry'] = 'ellipsoid'
df['label'] = np.where(df['Geometry'] == 'ellipsoid', 0, 1)
df
```

```python
degrees = []
s11 = []
s21_11 = []
for i in range(len(df)):
    with open(df['folder_path'][i]+'/mueller', 'r') as f:
        ncols = len(f.readline().split(' ')) - 1
        data = np.fromiter((v for line in f for v in line.split(' ')[:-1]), float ,count=-1)
        cols = data.reshape(int(data.size / ncols), ncols).transpose()
        degrees.append(cols[0])
        s11.append(cols[1])
        s21_11.append(cols[2]/cols[1])
        
df['degrees'] = degrees
df['S11'] = s11
df['-S21/S11'] = s21_11
```
